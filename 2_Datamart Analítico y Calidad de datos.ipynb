```python
#Función para convertir las fechas a datetime de manera rápida

def datetime_fast(dt, formato): 
    def divide_date(fecha):
        division = fecha.split()
        date= division[0]
        time = division[1]
        cadena = date + " " + time
        return(cadena)
    resultado = pd.to_datetime(dt.apply(lambda x: divide_date(x)), format=formato)
    return (resultado)
        
        
```

```python
df.event_time = datetime_fast(df.event_time, formato="%Y-%m-%d %H:%M:%S")
```

# Transformacion y creacion de variables

- Extraer componentes fecha
- Variables de calendario: festivos locales...
- Indicadores exógenos (ofertas competencia, black friday, san valentin...)


```python
def componentes_fecha(df):
    date = df.index.date
    año = df.index.year
    mes = df.index.month
    dia = df.index.day
    hora = df.index.hour
    minuto = df.index.minute
    segundo = df.index.second

    return(pd.DataFrame({"date":date , "año":año, "mes":mes, "dia":dia, "hora":hora, "minuto":minuto, "segundo":segundo}))
     
```

```python
df = pd.concat([df.reset_index(), componentes_fecha(df)], axis=1).set_index("fecha")
df
```
